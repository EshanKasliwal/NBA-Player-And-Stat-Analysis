# Exploring Player Demographics 

We looks at the-> 1. Height distribution of NBA players <br>
                  2. The Average height of NBA player <br>
                  3. Top 5 Tallest and Shortest Players in NBA <br>
                  Similarity for the Weight

```python
roster.head()
```

```python
# we need to change the height from feet to cm
def feet_to_cm(x):
    feet,inches = map(int,x.split('-'))
    return (feet * 30.48)+ (inches *2.54)
```

```python
roster['height_cm'] = roster['Ht'].apply(feet_to_cm)
```

```python
roster.head() #jConverted height into CM
```

1. Height Distribution 

```python
roster['height_cm'].hist(edgecolor='black')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency of players')
plt.title('Histogram of Height')
```

```python
roster['height_cm'].describe()
```

2. Average Height of NBA player 

```python
average_height = roster['height_cm'].mean()

print(f"The Average Height is: {average_height} cm")
```

3. Top 5 Tallest and shortest players in NBA 

```python
top5_tallest = roster.sort_values(by='height_cm', ascending=False).drop_duplicates(subset='Player').head(5)
top5_shortest = roster.sort_values(by='height_cm', ascending=True).drop_duplicates(subset='Player').head(5)

fig, ax = plt.subplots()

# Plotting the top 5 tallest individuals
ax.bar(top5_tallest['Player'], top5_tallest['height_cm'], color='blue', label='Top 5 Tallest')

# Plotting the top 5 shortest individuals
ax.bar(top5_shortest['Player'], top5_shortest['height_cm'], color='orange', label='Top 5 Shortest')


ax.set_xlabel('Name')
ax.set_ylabel('Height (cm)')
ax.set_title('Top 5 Tallest and Shortest Individuals')
ax.tick_params(axis='x', rotation=45)
ax.legend()
plt.show()
```

1.Looking at WEIGHT Distributions->

```python
roster['Wt'].hist(edgecolor='black')
plt.xlabel('Weight (pounds)')
plt.ylabel('Frequency of players')
plt.title('Histogram of Weight')
```

2. Average Weight of an NBA Player

```python
average_weight = roster['Wt'].mean()

print(f"The Average Weight is: {average_weight} pounds")
```

3. Top 5 Heaviest Players 

```python
top5_heaviest = roster.sort_values(by='Wt', ascending=False).drop_duplicates(subset='Player').head(5)

fig, ax = plt.subplots()
ax.bar(top5_heaviest['Player'], top5_heaviest['Wt'], color='blue', label='Top 5 Heaviest')
ax.set_xlabel('Name')
ax.set_ylabel('Weight (pounds)')
ax.set_title('Top 5 Heaviest Players')
ax.tick_params(axis='x', rotation=45)
ax.legend()
plt.show()
```

# Player Performance Analysis

```python
#extracting the 'season' information from the 'seasons' column
total['Season'] = total['season'].str.extract(r'(\d{4})')[0] + '-' + (total['season'].str.extract(r'(\d{2})')[0].astype(int) + 1).astype(str)
```

->Analyzing Points Per Game (PPG) for a specific player-  PPG calculation-> total points scored  / total number of games played

-We create a list of players to analyze 

-You can replace the players in the list with the ones you want to analyze, these 5 are my favorites:)  

```python
players_to_analyze = ['Michael Jordan', 'LeBron James','Stephen Curry','Giannis Antetokounmpo','Kevin Durant']
```

Creatings plots for each player 

```python
for player_name in players_to_analyze:
    player_data = total[total['Player'] == player_name].copy()
    player_data['PPG'] = player_data['PTS'] / player_data['G']    

    plt.figure(figsize=(10,4))
    plt.plot(player_data['Season'], player_data['PPG'], marker = 'o', linestyle = '-', label = player_name)
    plt.title('Points Per Game Over Time')
    plt.xlabel('Season')
    plt.ylabel('PPG')
    plt.xticks(rotation= 45)
    plt.grid(True)
    plt.legend(loc='upper left')
    plt.show()
```

# Top Performance in Playoffs 

```python
total_playoffs['Season'] = total_playoffs['season'].str.extract(r'(\d{4})')[0] + '-' + (total_playoffs['season'].str.extract(r'(\d{2})')[0].astype(int) + 1).astype(str)
```

```python
total_playoffs.head() #added a Season column with '-'
```

Creating merged data from total_playoffs & average_playoffs to include player statistics and then adding suffixes 

```python
merged_data = total_playoffs.merge(average_playoffs[['Player', 'PTS', 'TRB', 'AST']], on='Player', suffixes=('_total', '_average'))
```

```python
merged_data.head()
```

Grouping data by 'Player' and 'Season' to calculate total points, rebound and assists during playoffs 

```python
player_playoff_stats = merged_data.groupby(['Player', 'Season'])[['PTS_average','TRB_average', 'AST_average']].sum().reset_index()
```

Visualizing the Top 5 Performers in terms of points during playoff seasons  

```python
top_players_by_points = player_playoff_stats.groupby('Season')[['Player', 'PTS_average']].apply(lambda x: x.nlargest(5, 'PTS_average')).reset_index()
```

```python
top_players_by_points= top_players_by_points.sort_values(by='PTS_average', ascending=False)
plt.figure(figsize=(10,8))
for season,group in top_players_by_points.groupby('Season'):
    bars=plt.bar(group['Player'], group['PTS_average'], label= season)

    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() /2, bar.get_height(), str(int(bar.get_height())), ha='center', va= 'bottom')
plt.title('Top Players by Points in Playoffs')
plt.xlabel('Player')
plt.ylabel('Total Points')
plt.xticks(rotation= 45, ha= 'right')
plt.tight_layout()
plt.show()        
```

Calculating Top Performers in terms of rebounds and assists separately

```python
player_playoff_stats = merged_data.groupby(['Player', 'Season'])[['TRB_average','AST_average']].sum().reset_index()
```

```python
top_players_by_rebounds = player_playoff_stats.groupby('Season')[['Player', 'TRB_average']].apply(lambda x: x.nlargest(5, 'TRB_average')).reset_index()
top_players_by_assists = player_playoff_stats.groupby('Season')[['Player', 'AST_average']].apply(lambda x: x.nlargest(5, 'AST_average')).reset_index()
```

```python
plt.figure(figsize=(10,8))
for season,group in top_players_by_rebounds.groupby('Season'):
    bars=plt.bar(group['Player'], group['TRB_average'], label= season)

    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() /2, bar.get_height(), str(int(bar.get_height())), ha='center', va= 'bottom')
plt.title('Top Players by Rebounds in Playoffs')
plt.xlabel('Player')
plt.ylabel('Total Rebounds')
plt.xticks(rotation= 45, ha= 'right')
plt.tight_layout()
plt.show()        
```

```python
plt.figure(figsize=(10,8))
for season,group in top_players_by_assists.groupby('Season'):
    bars=plt.bar(group['Player'], group['AST_average'], label= season)

    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() /2, bar.get_height(), str(int(bar.get_height())), ha='center', va= 'bottom')
plt.title('Top Players by Assists in Playoffs')
plt.xlabel('Player')
plt.ylabel('Total Assists')
plt.xticks(rotation= 45, ha= 'right')
plt.tight_layout()
plt.show()        
```

Calculating total points during playoffs and regular season

```python
player_playoff_stats = merged_data.groupby(['Player', 'Season'])[['PTS_total','PTS_average']].sum().reset_index()
```

```python
#Visualizing top 15 performers in terms of points during playoff season
top_players_by_points = player_playoff_stats.groupby('Season')[['Player', 'PTS_total']].apply(lambda x: x.nlargest(15, 'PTS_total')).reset_index().head(15)
```

```python
colors = ['b','g','r','c','m','y','k']
plt.figure(figsize=(10,4))
for season, group in top_players_by_points.groupby('Season'):
    players= group['Player']
    points = group['PTS_total']
    color = colors.pop(3)
    bars = plt.bar(players,points,label=season,color=color)
    for bar, point in zip(bars, points):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(int(point)),ha = 'center', va = 'bottom')
plt.title('Top 15 Players by Points in Playoffs')
plt.xlabel('Player')
plt.ylabel('total points (playoffs)')
plt.xticks(rotation=45, ha= 'right')
plt.legend()
plt.show()
```

# Comparing regular-season and playoff performance for players

```python
comparison_data = player_playoff_stats.groupby('Player')[['PTS_total', 'PTS_average']].mean().reset_index()
comparison_data = comparison_data.sort_values('PTS_total', ascending=False).head(15) #selecting top 15
```

```python
plt.figure(figsize=(10,6))
bar_width = 0.35

bar1 = plt.bar(comparison_data['Player'], comparison_data['PTS_total'], label= 'Playoffs', alpha= 0.7, width=bar_width)
bar2 = plt.bar(comparison_data['Player'], comparison_data['PTS_average'], label= 'Regular Season', alpha= 0.7, width=bar_width)

plt.title('Comparison of Regular Season and Playoff Performance')
plt.xlabel('Player')
plt.ylabel('Average Points')
plt.xticks(rotation=45, ha="right")
plt.legend()

for bar1, bar2 in zip(bar1,bar2):
    plt.text(bar1.get_x() + bar1.get_width() /2, bar1.get_height(), str(int(bar1.get_height())), ha = 'center', va = 'bottom')
    plt.text(bar2.get_x() + bar2.get_width() /2, bar2.get_height(), str(int(bar2.get_height())), ha = 'center', va = 'bottom')

plt.show()
```
